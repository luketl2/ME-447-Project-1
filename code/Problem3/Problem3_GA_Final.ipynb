```python
import sys
sys.path.append('../')
from common.GeneticAlgorithm import GeneticAlgorithm
import numpy as np
import math

def fitness_func(total_pop):
    """ Calculates fitness using the given function f(x) = 10 *x^2
    Returns a (POP_SIZE, ) numpy array
    """       
    fitness = 10 * np.square(total_pop)
    return fitness.flatten()

def environmental_constraint(pop):
    return pop
```

```python
PARAM_SIZE = 1
POP_SIZE = 100
LOW_BOUND = -10
HIGH_BOUND = 10
NUM_GENERATIONS = 100
genetic_alg = GeneticAlgorithm(
    PARAM_SIZE, 
    POP_SIZE, 
    LOW_BOUND, 
    HIGH_BOUND, 
    fitness_func,
    environmental_constraint
)
genetic_alg.crossover_idx = 0
genetic_alg.run(NUM_GENERATIONS)
```
